```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.fft import * 
import sounddevice as sd

import librosa
import librosa.display
```

```python
sound, sr = librosa.load("./guitar-single-note.wav")

sound = sound[: (sr * 3) + 1]
print(f"Sound clip is {len(sound)/ sr} seconds long")

```

```python
plt.figure(figsize=(14, 5))

time = np.array(range(0, len(sound))) / sr

plt.plot(time[:1000], sound[:1000])
plt.xlabel("seconds")
plt.ylabel("amplitude")
```

```python
transform = fft(sound)

frequencies = np.array(range(0, len(transform))) / 3


plt.figure(figsize=[10,5])
plt.plot(frequencies, np.abs(transform))

condition = (frequencies > 100) & (frequencies < 1000)


idx = np.argmax(np.abs(transform))
print(frequencies[idx])

plt.figure(figsize=[10,5])
plt.plot(frequencies[condition], np.abs(transform[condition]))
plt.xlabel("frequencies (Hz)")

plt.axvline(x = frequencies[idx], color= 'red')
```
